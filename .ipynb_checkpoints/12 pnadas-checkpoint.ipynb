```python
import pandas as pd 
```

```python
# Define data using a dictionary
data = {'name': ['Jack', 'Sara', 'Sam'],
        'sex': ['male', 'female', 'male'],
        'age': [23, 25, 19],
        'address': [123, 256, 564]}

df = pd.DataFrame(data, index=[12, 13, 15]) # Create a DataFrame named 'df' with the data and specified index

# Display the DataFrame
df
```

```python
# Define data for a Series
ages = pd.Series([23, 25, 19], name='Age')

# Display the Series
ages
```
